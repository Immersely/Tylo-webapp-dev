````python
import requests

url_arxiv = "http://34.171.46.125:8080/query"
url_pubmed = "http://34.122.155.122:5000/query"
url_google = "http://34.42.208.54:4000/query?"
question = "entanglement entropy"

params = {"question": question}


try:
    response_arxiv = requests.get(url_arxiv, params=params)
    if response_arxiv.status_code == 200:
        print("Response from server:", response_arxiv.text)
    else:
        print("Error:", response_arxiv.status_code)
except requests.exceptions.RequestException as e:
    print("HTTP Request failed:", e)


try:
    response_pubmed = requests.get(url_pubmed, params=params)
    if response_arxiv.status_code == 200:
        print("Response from server:", response_pubmed.text)
    else:
        print("Error:", response_pubmed.status_code)
except requests.exceptions.RequestException as e:
    print("HTTP Request failed:", e)

try:
    response_google = requests.get(url_google, params=params)
    if response_google.status_code == 200:
        print("Response from server:", response_google.text)
    else:
        print("Error:", response_google.status_code)
except requests.exceptions.RequestException as e:
    print("HTTP Request failed:", e)
````
